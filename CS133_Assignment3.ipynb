```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


penguins = sns.load_dataset('penguins') # loading penguin dataset from seaborn
print("First few rows of penguin dataset: ")
penguins.head()





```

```python
#using loc and iloc to access specifc rows and columns
print("using loc to select rows and columns of species and island: ")
print(penguins.loc[0:4, ['species', 'island']])

print("using loc to select rows and columns of adelie and some columns: ")
print(penguins.loc[penguins['species']=='Adelie', ['bill_length_mm', 'flipper_length_mm']])

print("using iloc to select first 7 rows and 3 rows")
print(penguins.iloc[0:7, 0:4])

#spliting the island column into two seprate columns
print("splitting the island column into two seprate columns")
penguins['island_code'] = penguins['island'] + "-code" #creating a new column with delimiter
penguins_split = penguins['island_code'].str.split('-', expand=True) #splitting the column into two columns
print(penguins_split)

print("number of unique values in each column")
print(penguins.nunique())

print("checking for missing values")
print(penguins.isnull().sum())

```
