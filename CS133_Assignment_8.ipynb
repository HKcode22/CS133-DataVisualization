```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

dataset = 'https://raw.githubusercontent.com/csbfx/advpy122-data/master/world_happiness_report_2020.csv'

df = pd.read_csv(dataset)
df.head()


#regression plot between GDP per capita and Happiness Score
sns.regplot(x='Log_GDP_per_capita', y='Positive_affect', data=df)

#adding titles and labels
plt.title('Happiness Score vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')

plt.show()
```
