```python
import pandas as pd

url = 'https://gist.githubusercontent.com/clairehq/79acab35be50eaf1c383948ed3fd1129/raw/407a02139ae1e134992b90b4b2b8c329b3d73a6a/winemag-data-130k-v2.csv'
df = pd.read_csv(url)
df.head()
```

```python
wine_data = df.dropna(subset=['points', 'price'])

def points_to_rating(points):
    if points >= 95:
        return 5
    elif points >= 90:
        return 4
    elif points >= 85:
        return 3
    elif points >= 80:
        return 2
    else:
        return 1

wine_data['stars'] = wine_data['points'].apply(points_to_rating)
wine_data.head()

# mean_price = wine_data['price'].mean()
# wine_data['price_diff'] = wine_data['price'].apply(lambda x: x - mean_price)
# # wine_data.head()

#mean for price for 5,4,3,2 stars
mean_price_5stars = wine_data[wine_data['stars'] == 5]['price'].mean()
mean_price_4stars = wine_data[wine_data['stars'] == 4]['price'].mean()
mean_price_3stars = wine_data[wine_data['stars'] == 3]['price'].mean()
mean_price_2stars = wine_data[wine_data['stars'] == 2]['price'].mean()

print(mean_price_5stars)
print(mean_price_4stars)
print(mean_price_3stars)
print(mean_price_2stars)
```
